23. (3 pts) Write a function that takes in any word and counts the number of a particular letter in it. Give the function a default value if no word is passed in.

```python

def count_letter(word, letter):
    count = 0
    for char in word:
        if char == letter:
            count += 1
    return count
#example of how to use it
result = count_letter("hannahbanana", "a")
print(result)  # i just went with that because of my partners nickname, which should give 5. this took me a while to figure out.
```
